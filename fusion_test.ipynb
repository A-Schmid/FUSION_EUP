```python
import time
import sys
from FUSION import DHT11
from FUSION.visualize import Plot
import numpy as np

dht = DHT11(node_name = "dht", node_location="1104")
plot = Plot(title = "Humidity")

def on_humidity(data):
    plot.show(data)
    plot.update()

dht.OnUpdate(on_humidity, "humidity")
```
