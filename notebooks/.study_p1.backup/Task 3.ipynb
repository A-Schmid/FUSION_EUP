```python
import time
import sys
sys.path.append("../..")
from FUSION import Pin, BH1750
```

```python
led = Pin(node_name="led", pin_id=16, node_location="1104")
bh1750 = BH1750(node_name = "light_sensor", node_location="1104")

led.digitalWrite(1)

def on_light(data):
    #print("{} Einheiten Lichtintesit√§t".format(data))
    if data < 100:
        print('Dark')
        led.digitalWrite(1)
    else:
        print('Light')
        led.digitalWrite(0)
    
        
bh1750.OnUpdate(on_light, "light_intensity")

time.sleep(60)
```
