```python
from FUSION import *
import time

pirEntrance = pir(42)
pirRoom = pir(41)
pirDoor = pir(40)
pirWorkshop = pir(43)

def LogEvent(eventString):
    print(eventString)
    log_append_line(filename = "vr4_log.txt", message = eventString)

def OnEntranceEvent(event, timestamp):
    LogEvent("{} - {}: {}".format("entrance", event, time.ctime(timestamp)))

def OnRoomEvent(event, timestamp):
    LogEvent("{} - {}: {}".format("room", event, time.ctime(timestamp)))

def OnDoorEvent(event, timestamp):
    LogEvent("{} - {}: {}".format("door", event, time.ctime(timestamp)))

def OnWorkshopEvent(event, timestamp):
    LogEvent("{} - {}: {}".format("workshop", event, time.ctime(timestamp)))

    
pirEntrance.OnEvent(OnEntranceEvent)
pirRoom.OnEvent(OnRoomEvent)
pirDoor.OnEvent(OnDoorEvent)
pirWorkshop.OnEvent(OnWorkshopEvent)
```
